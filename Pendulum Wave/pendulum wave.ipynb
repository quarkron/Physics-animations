```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
times = [0, 1/6, 1/4, 1/2, 3/4, 5/6, 1] 
pendula=6
gamma = 20.5 
N = 20 
n_conts=np.linspace(0,pendula-1,200) 
i=0
def f(n,t, gamma = 20.5, N=20):
    return np.cos((2*np.pi)*(n+N)*t)
for t in times: 
    points = [f(n,t) for n in range(pendula)] 
    curve = f(n_conts,t)
    plt.figure(dpi=100)
    plt.scatter(range(0,pendula), points, c='r') 
    plt.plot(n_conts, curve, 'k--', linewidth=0.5)
    plt.ylim((-1.1,1.1))
    plt.xlabel('Pendulum Index') 
    plt.ylabel('Amplitude')
    plt.axhline(y=0, c='k') 
    time = str(round(t,3))
    plt.title(r't='+ time + r'$\Gamma$') 
    # plt.savefig('pendulum_wave_'+str(i) + '.png', dpi=300) 
    i+=1
```

# Damped Pendulum Wave Demonstration 

Note: Set decay_factor = 0 to demonstrate undamped pendulum wave. 

gamma is the period of the pendulum wave, and N is the number of oscillations of the zeroth-index pendulum after t=gamma.

```python
from matplotlib.animation import FuncAnimation 

plt.style.use('dark_background') 

gamma = 20 
N = 20 

def f(n,t, gamma,N, decay_factor=3):
    return np.cos((2*np.pi)*(n+N)*t)*np.exp(-decay_factor*t)

fig,ax = plt.subplots(figsize=(20,10))
scatter= ax.scatter([],[])
plt.xticks([])
plt.yticks([])

upper_bound=10
pendula=10
fps=50
frames_total = 1000
n_conts = np.linspace(0, upper_bound-1, pendula)
def update(frame):
    # Clear previous scatterplot
    ax.clear()

    # Generate new data for scatterplot
    points = [f(n,frame/frames_total, gamma=gamma, N=N) for n in n_conts]

    # Update scatter plot with new data
    scatter= ax.scatter(n_conts, points, s=100, c=np.linspace(0,1,pendula))

    #Set y-lims 
    plt.ylim((-1.1,1.1))

    plt.xticks([])
    plt.yticks([])


    # Return updated scatter plot
    return scatter,

# Create animation object
ani = FuncAnimation(fig, update, frames=frames_total, interval=1000/fps, blit=True)

ani.save('damped_pendulum_wave.gif', writer='pillow')  

# Show animation
plt.show()
```
